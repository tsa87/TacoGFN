```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import Descriptors
```

```python
zinc250k = pd.read_csv('../dataset/zinc_250k.csv')
```

```python
smiles = zinc250k['smiles'].to_list() #.sample(100).to_list()
smiles = [s.strip() for s in smiles]
mols = [Chem.MolFromSmiles(s) for s in smiles]
```

```python
# get molecular weight
mw = [Chem.Descriptors.MolWt(m) for m in mols]
```

```python
sum([True if m >= 160 and m <= 480 else False for m in mw]) / len(mw)
```

```python
with open('../dataset/zinc_rand_100.smi', 'w') as f:
    f.write('\n'.join(smiles))
```

```python
Draw.MolsToGridImage(mols, molsPerRow=5, subImgSize=(200, 200))
```
